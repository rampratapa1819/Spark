```
orderspath='/FileStore/tables/orders'
productspath='/FileStore/tables/products'
order_itemspath='/FileStore/tables/order_items'
departmentspath='/FileStore/tables/departments'
categoriespath='/FileStore/tables/categories'
customerspath='/FileStore/tables/customers'
```

```
ordersRdd=sc.textFile(orderspath)
```

```
ordersRdd.take(10)
```

```
# 1. Count orders by status
ordersmap=ordersRdd.map(lambda x:x.split(",")[3]).map(lambda y:(y,1)).reduceByKey(lambda x,y:x+y)
ordersmap.collect()
```

```
# 2. Percentage of closed orders
ordercountclosed=ordersRdd.filter(lambda x:x.split(',')[3]=='CLOSED')
ordercountall=ordersRdd.map(lambda x:x.split(",")[3])
(ordercountclosed.count()/ordercountall.count())*100
```

```
# 3. Average revenue per order
orderclosed=ordersRdd.filter(lambda x:x.split(',')[3]=='CLOSED')
ordercomplete=ordersRdd.filter(lambda y:y.split(',')[3]=='COMPLETE')
orderclosedprices=orderclosed.map(lambda x:int(x.split(',')[2]))
ordercompleteprices=ordercomplete.map(lambda x:int(x.split(',')[2]))
orderpricesum=orderclosedprices.sum()+ordercompleteprices.sum()
orderscount=orderclosed.count()+ordercomplete.count()
ordersRevenue=orderpricesum/orderscount
ordersRevenue
```
